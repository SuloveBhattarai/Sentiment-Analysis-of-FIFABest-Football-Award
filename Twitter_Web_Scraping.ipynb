```python
import pandas as pd
from ntscraper import Nitter
```

```python
scraper = Nitter()
```

```python
tweets = scraper.get_tweets("FIFATheBest", mode = 'term', number=1000, since="2024-01-17", until="2024-01-19")
```

```python
import pandas as pd
final_tweets = []
for x in tweets['tweets']:
    data =  x['text'],x['date']
    final_tweets.append(data)
```

```python
dat = pd.DataFrame(final_tweets, columns =['text','date'])
```
